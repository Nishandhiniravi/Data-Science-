```python
df_features = df[['Movie_Title','Movie_Genre','Movie_Language','Movie_Cast','Movie_Director']].fillna('')
df_features.shape
```

```python
x = df_features['Movie_Title'] + ' ' + df_features['Movie_Genre'] + ' ' + df_features['Movie_Language'] + ' ' + df_features['Movie_Cast'] + ' ' + df_features['Movie_Director']
x
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
x = tfidf.fit_transform(x)
print(x)
```


***Get Similarity Score using Cosine Similarity***

```python
from sklearn.metrics.pairwise import cosine_similarity
Similarity_Score = cosine_similarity(x)
Similarity_Score
Similarity_Score.shape
```

***Get Movie Name as Input from User and Validate for closest Spelling***

```python
Favourite_Movie_Name = input('Enter your favourite movie name : ')
All_Movies_Title_List = df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_Recommendation = difflib.get_close_matches(Favourite_Movie_Name, All_Movies_Title_List)
print(Movie_Recommendation)
```

```python
Close_Match = Movie_Recommendation
print(Close_Match)
```

```python
# Define the title you're searching for (string from user/input/UI)
query_title = "The Matrix"   # <-- put your title here (was user_title in my example)

from difflib import get_close_matches
import pandas as pd
import numpy as np

# Ensure titles are strings
titles = df['Movie_Title'].astype(str).tolist()

candidates = get_close_matches(query_title, titles, n=1, cutoff=0.6)

if not candidates:
    raise ValueError(f"No close match found for {query_title!r}")

close_match = candidates[0]

mask = df['Movie_Title'].astype(str).eq(close_match)
ids = df.loc[mask, 'Movie_ID']

if ids.empty:
    raise ValueError(f"Matched title {close_match!r} not present in df['Movie_Title'].")

Index_of_Close_Match_Movie = ids.iloc[0]
print(Index_of_Close_Match_Movie)
```

```python
Recommendation_Score = list(enumerate(Similarity_Score[Index_of_Close_Match_Movie]))
print(Recommendation_Score)
```

```python
len(Recommendation_Score)
```

```python
Sorted_Similar_Movies = sorted(Recommendation_Score, key = lambda x:x[1], reverse = True)
print(Sorted_Similar_Movies)
```

```python
print('Top 10 Movies Suggested for you : \n')
i = 1
for movie in Sorted_Similar_Movies:
  index = movie[0]
  title_from_index = df[df.index==index]['Movie_Title'].values[0]
  if (i<31):
    print(i, '.',title_from_index)
    i+=1
```
