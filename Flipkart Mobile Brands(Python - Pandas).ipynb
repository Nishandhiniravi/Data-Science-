```python
# 9. Find the count of Null values:
missing_values = data.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
# 10. Fill Null values in Rating with mean value
data['Rating'] = data['Rating'].fillna(data['Rating'].mean())
print(data)
```

```python
# 11. Count Number of duplicate rows:
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# 12. Drop Duplicate values
data = data.drop_duplicates()
data
```

```python
# 13. Total and average price of selling Price and count of the Models
aggregate = data.groupby(['Brand', 'Model']).agg(
    total_Price=('Selling Price', 'sum'),
    avg_Price=('Selling Price', 'mean'),
    Brand_count=('Brand', 'sum')
)

print(aggregate)
```

```python
# 14. Grouping Color and Brand
grouping = data.groupby(['Color', 'Brand'])
print(grouping.first())
```

```python
# 15. Maximum and minimum price per brand
price_status = data.groupby('Brand')['Selling Price'].agg(['min', 'max']).sort_values(by='max', ascending=False)
print(price_status)
```

```python
# 16. How many models does each brand
models_per_brand = data.groupby('Brand')['Model'].nunique().sort_values(ascending=True)
print(models_per_brand)
```

```python
# 17.Average rating per Memory size
avg_rating = data.groupby('Memory')['Rating'].mean().sort_values(ascending=False)
print(avg_rating)
```

```python
# 18. Total number of phones and average price by brand and storage
Total_Phones_AvgPrice = data.groupby(['Brand', 'Storage']).agg(
    Total_Models=('Model', 'nunique'),
    Avg_Price=('Selling Price', 'mean')
).sort_values(by='Avg_Price', ascending=False)

print(Total_Phones_AvgPrice)
```

```python
# 19. Average Price by brand
avg_price = data.groupby('Brand')['Selling Price'].mean().sort_values(ascending=False)
print("Average price by brand:\n", avg_price)
```

```python
#  Model Sorted
Models = data.sort_values
Models
```

```python
# 20. Top 5 Most Expensive Phones
Expensive=data.sort_values(by='Selling Price', ascending=False).head(5)
print("Top 5 Most Expensive Phones:\n", Expensive[['Brand', 'Model', 'Selling Price']])
```

```python
# 21. Top 5 Most Cheapest Phones
Cheapest=data.sort_values(by='Selling Price', ascending=True).head(5)
print("Top 5 Most Cheapest Phones:\n", Cheapest[['Brand', 'Model', 'Selling Price','Rating']])
```

```python
# 22. Trending Brand with Selling Price, Rating:
Trend = data.groupby('Brand')[['Selling Price', 'Rating']].mean().sort_values(by='Selling Price', ascending=False)
print("Price & Rating Trends by Brand:\n", Trend)
```

```python
# 23. Filtering the brands and mean value of selling Price Less than 11000:
filtered = data.groupby('Brand').filter(lambda x: x['Selling Price'].mean() <= 11000)
print(filtered)
```
