```python
# 1. Models of all Apple mobiles
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Apple_data = data[data['Brand'] == 'Apple']
x = Apple_data['Model']       
y = Apple_data['Selling Price']        

plt.bar(x, y)
plt.xlabel('Apple Products')
plt.ylabel('Sales')
plt.title('Sales of Apple Products')
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.show()
```

```python
# 2. Count of Apple brand Mobiles in the dataset
Apple_count=len(data[data['Brand']=='Apple'])
plt.figure(figsize=(4, 4))
plt.bar('Apple', Apple_count, color='red')
plt.title('Count of Apple brand Mobiles')
plt.ylabel('Number of Mobiles')
plt.ylim(100, Apple_count)  
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# 3. Which brand mobiles are Black
Black_brands = data[data['Color'] == 'Black']['Brand'].value_counts()
plt.figure(figsize=(8, 6))
Black_brands.plot(kind='bar', color='black')
plt.title('Count of Black Color Mobiles by Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Black Mobiles')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 4.Count the value of Brands
import pandas as pd
import matplotlib.pyplot as plt
top_brands = data['Brand'].value_counts().head()

plt.scatter(top_brands.index, top_brands.values, color='r')
plt.xlabel('Brand')
plt.ylabel('Number of Products')
plt.title('Top 5 Brands by Product Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 5. Count the Color
Count_Color=data['Color'].value_counts().head()

plt.barh(Count_Color.index, Count_Color.values)
plt.xlabel('Count of Brands')
plt.ylabel('Number of Colors')
plt.title('Number of Colors per Brand')
plt.xticks()
plt.tight_layout()
plt.show()
```

```python
# 6. Count the Rating
import pandas as pd
import matplotlib.pyplot as plt
Ratings = data['Rating'].value_counts().head()

plt.barh(Ratings.index.astype(str), Ratings.values, color='hotpink')
plt.xlabel('Count')
plt.ylabel('Ratings')
plt.title('Top 5 Most Frequent Ratings')
plt.tight_layout()
plt.show()

```

```python
7.# Count the Memory 
Count_Memory=data['Memory'].value_counts().head()
plt.scatter(Count_Memory.index, Count_Memory.values)
plt.xlabel('Memory')
plt.ylabel('Number of Products')
plt.title('Number of Products per Memory')
plt.xticks()
plt.tight_layout()
plt.show()
```

```python
# 9.Find the count of Null values
Null_values = data.isnull().sum()

plt.figure(figsize=(8, 5))
Null_values.plot(kind='bar', color='skyblue')
plt.title('Null Values per Column')
plt.xlabel('Columns')
plt.ylabel('Number of Null Values')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 10.Fill Null values in Rating with mean value
data['Original_Rating'] = data['Rating'].copy()
data['Rating'] = data['Rating'].fillna(data['Rating'].mean())

plt.hist(data['Original_Rating'].dropna(), bins=8, alpha=0.5, label='Original', edgecolor='black')
plt.hist(data['Rating'],bins=10, alpha=0.5, label='After Filling', edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Ratings Before and After Filling Missing Values')
plt.legend()
plt.show()
```

```python
# 11. Count Number of duplicate rows:
import matplotlib.pyplot as plt
import pandas as pd
duplicates = data.duplicated().sum()
labels = ['Unique Rows', 'Duplicate Rows']
counts = [len(data) - duplicates, duplicates]

plt.bar(labels, counts, color=['green', 'red'])
plt.title('Duplicate vs Unique Rows in Dataset')
plt.ylabel('Number of Rows')
plt.tight_layout()
plt.show()
```

```python
# 12. Drop Duplicate values
data_after = data.drop_duplicates()
plt.figure(figsize=(8,5))
plt.hist(data_after['Brand'], color='yellow', edgecolor='k')
plt.title("Drop Duplicate values")
plt.xlabel('Brand')
plt.ylabel('drop duplicates')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# 13.Total and average price of selling Price and count of the Models
import pandas as pd
import matplotlib.pyplot as plt
aggregate = data.groupby(['Brand', 'Model']).agg(
    total_Price=('Selling Price', 'sum'),
    avg_Price=('Selling Price', 'mean'),
    model_count=('Model', 'count')
).reset_index()
top_models = aggregate.sort_values('total_Price', ascending=False).head(10)

plt.barh(top_models['Model'], top_models['total_Price'], color='skyblue')
plt.xlabel('Total Selling Price')
plt.title('Top 10 Models by Total Selling Price')
plt.tight_layout()
plt.show()
```

```python
# 14.Grouping Color and Brand
import pandas as pd
import matplotlib.pyplot as plt
grouping = data.groupby(['Color', 'Brand']).size().reset_index(name='count')
top_color_brand = grouping.sort_values('count', ascending=False).head(10)

plt.barh(top_color_brand['Color'] + ' - ' + top_color_brand['Brand'], top_color_brand['count'], color='purple')
plt.xlabel('Number of Products')
plt.title('Top 10 Color-Brand Combinations by Product Count')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# 15. Maximum and minimum price per brand
import pandas as pd
import matplotlib.pyplot as plt
price_status = data.groupby('Brand')['Selling Price'].agg(['min', 'max']).sort_values(by='max', ascending=False)

price_status['max'].plot(kind='bar', color='orange', figsize=(10, 6))
plt.ylabel('Max Selling Price')
plt.title('Max Selling Price by Brand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

price_status = data.groupby('Brand')['Selling Price'].agg(['min', 'max']).sort_values(by='max', ascending=False)

price_status['min'].plot(kind='bar', color='orange', figsize=(10, 6))
plt.ylabel('Min Selling Price')
plt.title('Min Selling Price by Brand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# 16. How many models does each brand
models_per_brand = data.groupby('Brand')['Model'].nunique().sort_values(ascending=True)
print(models_per_brand)
sns.histplot(data=models_per_brand, kde=True)
plt.title("How many models does each brand")
plt.show()
```

```python
# 17. Average rating per Memory size
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
avg_rating = data.groupby('Memory')['Rating'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 8))
sns.boxplot(data=avg_rating, hue='Memory', y='Rating', palette='dark')
plt.xticks(rotation=45)
plt.title('Average Rating by Memory')
plt.xlabel('Memory')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
# 18. Total number of phones and average price by brand and storage
Total_Phones_AvgPrice = data.groupby(['Brand', 'Storage']).agg(
    Total_Models=('Model', 'nunique'),
    Avg_Price=('Selling Price', 'mean')
).sort_values(by='Avg_Price', ascending=False)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=Total_Phones_AvgPrice, x='Brand', y='Storage', palette='coolwarm',hue='Avg_Price')
plt.xticks(rotation=45)
plt.title('Total number of phones and average price by brand and storage')
plt.xlabel('Memory')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
# 19. Average Price by brand
avg_price = data.groupby('Brand')['Selling Price'].mean().sort_values(ascending=False)
sns.histplot(data=avg_price, kde=True)
plt.title("Average Price by brand")
plt.show()
```

```python

20.# Top 5 Most Expensive Phones
Expensive = data.sort_values(by='Selling Price', ascending=False).head(5)

plt.figure(figsize=(5, 5))
sns.barplot(data=Expensive, x='Brand', y='Selling Price', palette='viridis',legend=False)
plt.title('Top 5 Most Expensive Products')
plt.xlabel('Brand')
plt.ylabel('Selling Price')
plt.xticks()
plt.tight_layout()
plt.show()
```

```python
21.# Top 5 Most Cheapest Phones
Cheapest=data.sort_values(by='Selling Price', ascending=True).head(5)

plt.figure(figsize=(5, 5))
sns.barplot(data=Cheapest, x='Brand', y='Selling Price', palette='viridis',legend=False)
plt.title('Top 5 Most Cheapest Products')
plt.xlabel('Brand')
plt.ylabel('Selling Price')
plt.xticks()
plt.tight_layout()
plt.show()
```

```python
# 22. Trending Brand with Selling Price, Rating:
Trend = data.groupby('Brand')[['Selling Price', 'Rating']].mean().sort_values(by='Selling Price', ascending=False)
plt.figure(figsize=(8, 5))
sns.heatmap(Trend.corr(), annot=True, cmap='YlGnBu', fmt='.1f')
plt.title('Avg Selling Price & Rating (Heatmap)')
plt.xlabel('Selling Price and Rating')
plt.ylabel('Brands')
plt.tight_layout()
plt.show()
```

```python
# 23. Filtering the brands and mean value of selling Price Less than 11000:
filtered = data.groupby('Brand').filter(lambda x: x['Selling Price'].mean() <= 11000)
sns.pairplot(filtered, hue='Brand', palette='pastel')
plt.suptitle('Filtered Brands (Avg Price â‰¤ 11000)', y=1.02)
plt.show()
```

```python
# 24.Filter the Selling Price greater than 150000
filtered = data[data['Selling Price'] > 150000]
plt.figure(figsize=(5, 5))
sns.histplot(filtered['Selling Price'], bins=5, kde=True, color='gray')
plt.title('Filter the Selling Prices > 150,000')
plt.xlabel('Selling Price')
plt.ylabel('Level')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()  
```

```python
25.# Filtering Rating greater than 4.4 list out 'Brand', 'Model','Rating':
filtered = data.loc[data['Rating'] > 4.4, ['Brand', 'Model','Rating']]
sns.histplot(filtered['Rating'], bins=5, kde=True, color='gray')
plt.title('Filtering Rating greater than 4.4')
plt.xlabel('Rating')
plt.ylabel('Level')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
