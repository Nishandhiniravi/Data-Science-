```python
data = data.sort_values(by='Date')
```

```python
daily_sales = data.groupby(['Date', 'Store'])['IsHoliday'].sum().reset_index()
```

```python
# Group by Date and Store
store_sales = data.groupby(['Date', 'Store'])['Weekly_Sales'].sum().reset_index()

# Choose a specific store for forecasting
store_1 = store_sales[store_sales['Store'] == 1]

# Set Date as index
store_1 = store_1.set_index('Date').asfreq('W').fillna(0)

# Preview
store_1.head()
```

```python
import matplotlib.pyplot as plt

store_1['MA_4'] = store_1['Weekly_Sales'].rolling(window=4).mean()  # 4-week MA

store_1[['Weekly_Sales', 'MA_4']].plot(figsize=(12,5), title='4-Week Moving Average')
plt.show()
```
